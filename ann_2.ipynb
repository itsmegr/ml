```python
class NeuralNetwork:
    def __init__(self,inp,hid,out):
        
       # print(str(inp)+" "+str(hid))
        self.hidden_weights=np.random.randn(inp,hid)
        #print(self.weights_hidden)
        self.output_weights=np.random.randn(hid,out)
        #self.hidden_weights=np.random.randn(inp,hid)
        
        self.hidden_bias=np.zeros(hid)
        self.output_bias=np.zeros(out)
        
        self.learning_rate=0.1
        
    def sigmoid(self,s,deriv=False):
        if(deriv):
            return s*(1-s)
        return 1/(1+np.exp(-s))
    def feed_forward(self,inputs):
        self.hidden_sigma=np.dot(inputs,self.hidden_weights)+self.hidden_bias
        self.hidden_output=self.sigmoid(self.hidden_sigma)
        #print(inputs.shape)
        
        self.output_sigma=np.dot(self.hidden_output,self.output_weights)+self.output_bias
        self.output=self.sigmoid(self.output_sigma)
        return self.output
    def back_propagation(self,inputs,desired_output):
        self.output_error=(desired_output-self.output)
        self.output_delta=self.sigmoid(self.output,deriv=True)*self.output_error
        
        self.hidden_error=np.dot(self.output_weights,self.output_delta.T).T
        self.hidden_delta=self.sigmoid(self.hidden_output,deriv=True)*self.hidden_error
        #print(self.hidden_output.T)
        self.output_weights+=self.learning_rate*np.dot(self.hidden_output.reshape(-1,1),self.output_delta.reshape(1,-1))
        self.hidden_weights+=self.learning_rate*np.dot(inputs.reshape(-1,1),self.hidden_delta.reshape(1,-1))
        
        self.output_bias+=self.learning_rate*(self.output_delta)
        self.hidden_bias=self.learning_rate*(self.hidden_delta)
    def train(self,inputs,desired_output):
        self.feed_forward(inputs)
        self.back_propagation(inputs,desired_output)
        
        
```

```python
nn=NeuralNetwork(784,350,10)

for i in range(y_train.shape[0]):
    #print(X_train[0][i])
    #print(X_train[i][0])
    nn.train(X_train[i],y_train[i])
 
```
